```python
import statsapi
import pandas as pd
import numpy as np
from datetime import date
```

```python
# Dates
start_date='07/23/2020'
todays_date = date.today()
output_dir = 'data/'
```

```python
# # Get team ids -- need them for future queries
# teams = statsapi.get('teams', {
#     'sportIds': 1
# }, force=False)['teams']

# team_lst = []
# for t in teams:
#     team_lst.append({ 
#         'name': t['name'],
#         'id': t['id'] 
#     })
# pd.DataFrame(team_lst).to_csv(output_dir + 'teams.csv', index=False)
team_df = pd.read_csv(output_dir + 'teams.csv')
```

```python
game_lst = []
for idx, row in team_df.iterrows():
    games = statsapi.schedule(start_date=start_date, end_date=todays_date, team=row['id'], opponent="", sportId=1, game_id=None)
    keys = ['game_id', 'game_date', 'status', 'home_id', 'away_id']
    for g in games:
        subset = {k: g.get(k, None) for k in keys}
        subset['team_name'] = row['name']
        # Determine outcome if game has been played
        if g['status'] == 'Final':
            subset['home_score'] = g['home_score']
            subset['away_score'] = g['away_score']
            subset['win_status'] = 1 if g['winning_team'] == row['name'] else 0
        else:
            subset['home_score'] = np.nan
            subset['away_score'] = np.nan
            subset['win_status'] = np.nan
        game_lst.append(subset)
```

```python
pd.DataFrame(game_lst).to_csv(output_dir + str(todays_date) + '.csv', index=False)
```
